```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
```

```python
scores_df = pd.read_csv("Resources/2013-to-2019-World-Class-DCI-scores.csv")
scores_df
```

```python
#Add a year column for easier filtering
years = []
for date in scores_df['Date']:
    date_arr = date.split("-")
    year = date_arr[0]
    years.append(year)
years

scores_df['year'] = years

scores_df
```

```python
def top_5(year):
    scores_year = scores_df.loc[scores_df['year'] == str(year), :]
    
    scores_year

    min_date = scores_year['Date'].min()

    max_date = scores_year['Date'].max()

    top_5_df = scores_year.loc[scores_year['Date'] == max_date]

    top_5_df = top_5_df.head()
    
    first_place = top_5_df.iloc[0,4]
    second_place = top_5_df.iloc[1,4]
    third_place = top_5_df.iloc[2,4]
    fourth_place = top_5_df.iloc[3,4]
    fifth_place = top_5_df.iloc[4,4]
    
    return(first_place, second_place, third_place, fourth_place, fifth_place, min_date, max_date)
```

```python
#2013 Place markers
first_place = top_5('2013')[0]
second_place = top_5('2013')[1]
third_place = top_5('2013')[2]
fourth_place = top_5('2013')[3]
fifth_place = top_5('2013')[4]
```

```python
#Narrow down full scores to 2013 only
scores_2013 = scores_df.loc[scores_df['year'] == '2013', :]
scores_2013

first_df = scores_2013.loc[scores_2013['Corps'] == first_place, :]
second_df = scores_2013.loc[scores_2013['Corps'] == second_place, :]
third_df = scores_2013.loc[scores_2013['Corps'] == third_place, :]
fourth_df = scores_2013.loc[scores_2013['Corps'] == fourth_place, :]
fifth_df = scores_2013.loc[scores_2013['Corps'] == fifth_place, :]
```

```python
#Plot all five 2013 Lines
plt.plot(first_df['Score'], label='Carolina Crown')
plt.plot(second_df['Score'], label='The Blue Devils')
plt.plot(third_df['Score'], label='The Cadets')
plt.plot(fourth_df['Score'], label='Santa Clara Vanguard')
plt.plot(fifth_df['Score'], label='The Bluecoats')
plt.xticks([0, 250, 525], ["Prelims", " ", "Finals"])
plt.legend()
plt.title('DCI 2013 Season')
plt.ylabel('Score')
```
